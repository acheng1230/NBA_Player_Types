```python
# Load packages
import re
import pandas as pd
import numpy as np

np.set_printoptions(suppress=True)
```

```python
test = pd.read_csv('/Users/alexcheng/Desktop/test_file.csv')
test.head()
```

```python
test.columns
```

```python
test.drop(['Rk', 'Tm', 'MP', 'Unnamed: 19', 'Unnamed: 24', 'PER'], axis=1, inplace=True)
```

```python
regex_list = []
name_list = []

player_list = list(test['Player'])

for player_id in player_list:
    this_regex = re.findall('\\\(\w+)', player_id)[0]
    regex_list.append(this_regex)
    this_name = re.findall('\w.+\s\w+', player_id)[0]
    name_list.append(this_name)

test['Player_ID'] = regex_list
test['Player'] = name_list
```

```python
test
```

```python
test.shape
```

```python
groupby = test.groupby([test.Player, test.Pos, test.Player_ID]).mean()
```

```python
groupby.to_csv('/Users/alexcheng/Desktop/groupby.csv')
```
